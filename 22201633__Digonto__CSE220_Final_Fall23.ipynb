```python
def hashFunction(key,operations):
  # Write you code here
  import numpy as np
  new = np.array([None]*len(key))
  w = 0
  for i in key:
    for j in operations:
      if i == j[0]:
        new[w] = j[1]
        w += 1
    new[w]=i
    w+=1
  return new


KeyArr = hashFunction([1,2,4,6],[[1,3],[4,7],[6,1]])
print('The Updated key array is -- ', KeyArr)

KeyArr1 = hashFunction([1,2],[[1,3],[2,1],[3,2]])
print('The Updated key array is -- ', KeyArr1)
```

```python

# Binary Tree
class Node:
	def __init__(self, key):
		self.left = None
		self.right = None
		self.val = key


def printOrder(root):
	# Write your code here
	if root==None:
		return
	else:
		printOrder(root.right)
		printOrder(root.left)
		print(root.val,end=" ")

# Driver Code
root = Node('1')
root.left = Node('2')
root.right = Node('3')
root.left.left = Node('4')
root.left.right = Node('5')
root.right.left = Node('6')
root.right.right = Node('7')

printOrder(root)
```
